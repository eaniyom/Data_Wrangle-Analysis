## 1. Introduction
Data Wrangling could be seen as the process of cleaning and unifying messy and complex datasets for easy access and analysis. This can also be known as data munging, which involves the process of cleaning, organizing, and transforming raw data into desired format for proper decision making. 

Data Wrangling does not have a one way rule, it is usually dependent on the data being worked on. Research shows professionals spend about 80% of their time in data wrangling, while the remaining 20% can be used for exploration and modelling.  

The act of Data wrangling has been of great benefits to data analyst. Data wranging help does the following:
* Make raw data usable
* Centralizes data from several sources and location
* It acts as the preparation stage for data mining
* Cleansing of data from noise and flaws
* Helps for concrete business decisions

## 2. Steps Involved During Data Wrangling
The steps involved during the act of data wrangling can be summarized in to five. They include:
1. Data Gathering
2. Accessing Data
3. Data Cleaning
4. Data Storage
5. Others (Analysis & Visualizations)

These steps are not rigid, they could be changed depending on the data involved.

#### 1. Data Gathering 
Data gathering is also known as data collection. It is the process of gathering and measuring information on targeted variables in an established system. It involves the collection of data from several sources for the purpose of analyzing the data and driving insights from the data.

During Data wrangling, data gathering can be done in several ways and format. But for the purpose of this act, I will be limiting the my discussion to three basic method of gathering data, They include: 
1. Downloading Dataset manually (using in .csv  format)
2. Programmatically downloading of data, using the python request library
3. The Use of APIs to scrap data

These methods are solely dependent on the analyst involved, the nature of the data and the problem to be solved

#### 2 Accessing Data
Accessing Data is a step in the data wrangling act which involves the process of scanning, perusing and thoroughly checkin the datasets gathered. The data is being check to sort for errors and other irregularities related to the data. Accessing of data can be done:
* Visually - involves scrolling through the data to check for irregularities. &
* Programmatically - This involves the use of codes to check for errors associated with the data


We access data to check for the quality and how tidy the data is. This is subject to the persons involved in the data acessing process.

#### 3. Data Cleaning
After Accessing the data and fetching out the irregularities with the datasets, it is expedient that those issues be handled. This step is very important as it will be keep the quality of the datasets and keep the data tidy, and ready for storage and further analysis. The method of data cleaning is relative to the issues encountered and the analyst creativity and exposure

#### 4. Data Storage
At this stage, your data is ready to be stored for further analysis. The Storage of the data is where you select which table is to be added and which is not relevant. It is preferred the master data to be stored in data.cvs format for easy interaction with pandas DataFrame.
